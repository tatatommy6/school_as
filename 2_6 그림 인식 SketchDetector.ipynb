### [RoboCam 라이브러리 가져오기]

````python
from RoboCam.robocam import *
rCam = RoboCam()
````

### [ 스트리밍 준비/시작]

````python
rCam.WebcamStreamInit()
rCam.WebcamStream()
````

### [그림인식 초기화/실행]

````python
rCam.SketchDetectorInit()
rCam.SketchDetectorStart()
````

### [SketchDetector 함수]

- SketchDetectorInit() : 그림인식 모듈 준비(검은 외곽 사각형 내부에 그려진 그림을 학습하고 구별)
- SketchDetectorStart() : 그림인식 모듈 시작
- SketchDetectorStop() : 그림인식 모듈 정지
- SketchCapture(name:str,captureCount:int=5,path:str) : 인식된 그림을 name 이름으로 캡쳐하여 path에 저장
- TrainSketchData(sketchPath:str) : sketchPath에 저장된 그림 데이터를 학습
- DeleteSketchData(name:str, sketchPath:str) : sketchPath에 저장된 name 이름을 가진 그림 데이터 삭제(TrainFaceData 함수를 호출해야 적용)
- GetSketchNames() : 인식된 그림의 이름을 list형태로 반환

### [인식된 그림의 이름 출력하기]

- 인식된 그림의 이름이 리스트로 반환(리스트의 0번지 출력)
- 인식된 그림이 없으면 None 반환
- None 반환되는 경우 리스트의 0번지 출력하면 Error
- 학습되지 않은 이미지 : 'Sketch'라고 출력

````python
# 리스트 출력
print(rCam.GetSketchNames())
````

````python
# 리스트의 0번지 출력
print(rCam.GetSketchNames()[0])
````

````python
# 학습되지 않은 이미지 리턴값은 Sketch
print(rCam.GetSketchNames()[0])
````

````python
# 인식된 이미지 없으면 None 리턴
print(rCam.GetSketchNames())
````

````python
# 이미지 인식 안된경우 리스트의 0번지 출력하면 에러
print(rCam.GetSketchNames()[0])
````
