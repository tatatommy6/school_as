## [3] 동작 함수 정의하기

- 전진, 후진, 정지

### 1) 전진

image.png

```python
# 전진
def forward():
    rs.motor_write(0, 1, 5)
    rs.motor_write(1, 0, 5)
    rs.motor_write(2, 1, 5)
    rs.motor_write(3, 0, 5)
```

### 2) 후진

image.png

```python
# 후진
def reverse():
    rs.motor_write(0, 0, 5)
    rs.motor_write(1, 1, 5)
    rs.motor_write(2, 0, 5)
    rs.motor_write(3, 1, 5)
```

### 3) 정지

```python
# motor_stop()이용하여 정지
def stop():
    rs.motor_stop(0)
    rs.motor_stop(1)
    rs.motor_stop(2)
    rs.motor_stop(3)

# motor_write()이용하여 정지
def stop():
    rs.motor_write(0, 0, 0)
    rs.motor_write(1, 0, 0)
    rs.motor_write(2, 0, 0)
    rs.motor_write(3, 0, 0)
```

## [4] 함수 호출하여 움직여보기

```python
# 전진,후진,정지 테스트
forward()
time.sleep(0.5)
reverse()
time.sleep(0.5)
stop()
```

## [6] 동작 함수 속도 매개변수 넣기

```python
# 전진
def forward(speed=5):
    rs.motor_write(0, 1, speed)
    rs.motor_write(1, 0, speed)
    rs.motor_write(2, 1, speed)
    rs.motor_write(3, 0, speed)

# 후진
def reverse(speed=5):
    rs.motor_write(0, 0, speed)
    rs.motor_write(1, 1, speed)
    rs.motor_write(2, 0, speed)
    rs.motor_write(3, 1, speed)

# 정지
def stop():
    rs.motor_write(0, 0, 0)
    rs.motor_write(1, 0, 0)
    rs.motor_write(2, 0, 0)
    rs.motor_write(3, 0, 0)
    time.sleep(0.1)

# 우측
def right(speed=5):
    rs.motor_write(0, 1, speed)
    rs.motor_write(1, 1, speed)
    rs.motor_write(2, 0, speed)
    rs.motor_write(3, 0, speed)
    time.sleep(0.1)

# 좌측
def left(speed = 5):
    rs.motor_write(0, 0, speed)
    rs.motor_write(1, 0, speed)
    rs.motor_write(2, 1, speed)
    rs.motor_write(3, 1, speed)
    time.sleep(0.1)

#우회전
def right_turn(speed = 5):
    rs.motor_write(0, 1, speed)
    rs.motor_write(1, 1, speed)
    rs.motor_write(2, 1, speed)
    rs.motor_write(3, 1, speed)
    time.sleep(0.1)

#좌회전
def left_turn(speed = 5):
    rs.motor_write(0, 0, speed)
    rs.motor_write(1, 0, speed)
    rs.motor_write(2, 0, speed)
    rs.motor_write(3, 0, speed)
    time.sleep(0.1)
```

```python
#한쪽 바퀴 정지하여 좌회전,우회전
def RR(speed = 5):
    rs.motor_write(0, 1, speed)
    rs.motor_write(1, 0, 0)
    rs.motor_write(2, 1, speed)
    rs.motor_write(3, 0, 0)

def LL(speed = 5):
    rs.motor_write(0, 1, 0)
    rs.motor_write(1, 0, speed)
    rs.motor_write(2, 1, 0)
    rs.motor_write(3, 0, speed)
```

```python
forward(10)
time.sleep(0.5)
reverse(10)
time.sleep(0.5)
stop()
```

```python
right_turn(10)
time.sleep(0.5)
left_turn(10)
time.sleep(0.5)
stop()
```

```python
RR(10)
time.sleep(1)
LL(10)
time.sleep(1)
stop()
```
